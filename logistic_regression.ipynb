```python
crd_data = pd.read_csv("/content/credit_risk.csv")
```

```python
crd_data.info()
```

```python
crd_data.head()
```

```python
crd_data['class'].unique()
```

```python
x = crd_data.columns.drop("class")
y = crd_data['class']
```

```python
crd_data_enc = pd.get_dummies(crd_data[x])
crd_data_enc.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(crd_data_enc,y,test_size = 0.15,random_state = 100)
```

```python
print("Shape of X_train:", x_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", x_test.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
train_accuracy = model.score(x_train,y_train)
print("Train Accuracy =",train_accuracy)
test_accuracy = model.score(x_test,y_test)
print("Test Accuracy = ",test_accuracy)
```

```python
train_predictions = model.predict(x_train)
test_predictions = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
train_conf_matrix = confusion_matrix(y_train,train_predictions)
```

```python
pd.DataFrame(train_conf_matrix,columns=model.classes_,index=model.classes_)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,test_predictions))
```
